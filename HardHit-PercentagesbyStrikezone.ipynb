```python
import pandas as pd
import numpy as np
import pybaseball
import matplotlib.pyplot as plt
# load the Statcast data into a pandas DataFrame
statcast_data = pd.read_csv('brady.csv')
#Validate data, all 'launch_speed' should be >= 95  
statcast_data[['game_date','player_name','zone','pitch_type','release_speed','launch_speed','launch_angle','events',
               'description']]
```

```python
#Method to count the number of times the pitch went in the specific zone. Then divided by number of pitches times 100 to
#give us a percentage of the amount relative to the others
zone_1 = (statcast_data['zone'].value_counts()[1] / 94) * 100
zone_1 = "%.2f" % zone_1
zone_2 = (statcast_data['zone'].value_counts()[2] / 94) * 100
zone_2 = "%.2f" % zone_2
zone_3 = (statcast_data['zone'].value_counts()[3] / 94) * 100
zone_3 = "%.2f" % zone_3
zone_4 = (statcast_data['zone'].value_counts()[4] / 94) * 100
zone_4 = "%.2f" % zone_4
zone_5 = (statcast_data['zone'].value_counts()[5] / 94) * 100
zone_5 = "%.2f" % zone_5
zone_6 = (statcast_data['zone'].value_counts()[6] / 94) * 100
zone_6 = "%.2f" % zone_6
zone_7 = (statcast_data['zone'].value_counts()[7] / 94) * 100
zone_7 = "%.2f" % zone_7
zone_8 = (statcast_data['zone'].value_counts()[8] / 94) * 100
zone_8 = "%.2f" % zone_8
zone_9 = (statcast_data['zone'].value_counts()[9] / 94) * 100
zone_9 = "%.2f" % zone_9
zeleven = (statcast_data['zone'].value_counts()[11] / 94) * 100
zeleven = "%.2f" % zeleven
ztwelve = (statcast_data['zone'].value_counts()[12] / 94) * 100
ztwelve = "%.2f" % ztwelve
z13 = 0
z13 = "%.2f" % z13
zone_14 = (statcast_data['zone'].value_counts()[14] / 94) * 100
zone_14 = "%.2f" % zone_14
#Relative strikezone for each zone with percentages
print(zeleven + "                           " + ztwelve)
print("         " + zone_1 + "  " + zone_2 + "  " + zone_3)
print("         " + zone_4 + "  " + zone_5 + "  " + zone_6)
print("         " + zone_7 + "  " + zone_8 + "  " + zone_9)
print(z13 + "                           " + zone_14)

print("  ")
```

```python
#Pie chart percentages for each zone
zonedata = np.array([zone_1, zone_2, zone_3, zone_4, zone_5, zone_6, zone_7, zone_8, zone_9, zeleven, ztwelve, z13, zone_14])
zonelabels = ["Zone 1", "Zone 2", "Zone 3", "Zone 4", "Zone 5", "Zone 6", "Zone 7", "Zone 8", "Zone 9", "Zone 11", 
              "Zone 12", "Zone 13", "Zone 14"]
plt.pie(zonedata, labels = zonelabels)
plt.show() 
```
